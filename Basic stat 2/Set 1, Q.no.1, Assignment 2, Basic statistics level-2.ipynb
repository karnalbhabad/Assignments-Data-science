```python
#Pie plot
plt.figure(figsize=(18,9))
plt.pie(x,labels=Name,autopct='%1.0f%%')
plt.title('Name of the companies with respect to x', fontsize = 20, color ='red')
plt.show()
```

```python
#Finding Outliers using Boxplot
sns.boxplot(x)
```

```python
#Calculate mean
x.mean()
```

```python
#Calculate standard deviation(std)
x.std()
```

```python
#Calculate variance 
x.var()
```

```python
x = pd.Series(x)
Q1 = np.quantile(x,0.25)
Q3 = np.quantile(x,0.75)
IQR = Q3 - Q1
upper_bound = Q3+(1.5*IQR)
lower_bound = Q1-(1.5*IQR)
med = np.median(x)
print('First Quantile=',Q1, 'Second Quantile=',med, 'Third Quantile=',Q3, 'Inter Quantile Range=',IQR, 'Upper Whisker=',upper_bound, 'Lower Whisker Range=',lower_bound)
Outliers = x[(x<=lower_bound) | (x>=upper_bound)]
print('The Outlier in the boxplot:', Outliers)
```
